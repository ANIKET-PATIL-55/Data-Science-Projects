```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Check for missing values and handle them if necessary
titanic.dropna(subset=['age'], inplace=True)  # Remove rows with missing age values
```

```python
# Create a violin plot to show age distribution by survival
sns.violinplot(data=titanic, x='survived', y='age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
t_stat,p_value=ttest_ind(income_kp281,income_kp481)
p_value
if p_value<0.05:
    print(f'reject the null hypothesis, p value ={p_value}')
else:
    print(f'fail to reject the null hypothesis, p value ={round(p_value,3)}')
```
